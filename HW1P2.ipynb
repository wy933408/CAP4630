```
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
mpl.rc('axes', labelsize=14)
mpl.rc('xtick', labelsize=12)
mpl.rc('ytick', labelsize=12)
```

```
import numpy as np
np.random.seed(42)

x1 = 2 * np.random.rand(100, 1)
x2 = 4 * np.random.rand(100, 1)
yt = 4 + 3 * x1 + np.random.rand(100, 1) + 2 * x2 + np.random.rand(100, 1)
ys = yt.ravel()
```

```
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x1, x2, ys, 'b.')
ax.set_xlabel('x_1$')
ax.set_ylabel('x_2$')
ax.set_zlabel('y')
```

```
# split the data into training and test sets
# train set
train_x1 = x1[:80]
train_x2 = x2[:80]
train_ys = ys[:80]
# test set
test_x1 = x1[80:]
test_x2 = x2[80:]
test_ys = ys[80:]
```

# Linear regression using numpy


```
# number of epochs
epochs = 10
# learning rate
lr = 0.01
```

```
# initial value for weight w and bias b
w1 = np.random.randn(1)
w2 = np.random.randn(1)
b = np.zeros(1)
```

```
for epoch in np.arange(epochs):
  for i in np.arange(80):
    y_pred = w1 * train_x1[i] + b + w2 * train_x2[i]
    
    grad_w1 = (y_pred - train_ys[i]) * train_x1[i]
    grad_w2 = (y_pred - train_ys[i]) * train_x2[i]
    grad_b = (y_pred - train_ys[i])
    
    w1 -= lr * grad_w1
    w2 -= lr * grad_w2
    b -= lr * grad_b
```

```
test_loss = 0
for i in np.arange(20):
  test_loss += 0.5 * (w1 * test_x1[i] + b + w2 * test_x2[i]- test_ys[i]) ** 2
test_loss /= 20
```

```
pred_yt = w1 * test_x1 + b + w2 * test_x2
pred_ys = pred_yt.ravel()
```

```
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(test_x1, test_x2, pred_ys, 'b.')
ax.set_xlabel('x_1$')
ax.set_ylabel('x_2$')
ax.set_zlabel('y')
```
