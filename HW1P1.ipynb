```
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```
def f1(x,y):
  return y ** 2 * x ** 2
def f2(x,y):
  return -(y ** 2) - x ** 2
def f3(x,y):
  return y ** 3 + 3 * y ** 2  + x ** 3 + 3 * x ** 2
```

```
x = np.linspace(-6, 6, 50)
y = np.linspace(-6, 6, 50)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
```

```
plt.contour(X, Y, Z)
```

```
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```
x = np.linspace(-6, 6, 50)
y = np.linspace(-6, 6, 50)

X, Y = np.meshgrid(x, y)
Z = f2(X, Y)
```
